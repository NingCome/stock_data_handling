```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 从stock_data文件夹中读取sh600156.csv文件
df = pd.read_csv(r'stock_data\all_trading_data\stock data\sh600156.csv')
# 删掉其中的两列
df.drop(['code','report_type', 'report_date'], axis=1, inplace=True)
df.head()
```

```python

#创建10个df1的平移数据dataframe
data_x = pd.concat([df.shift(-i).add_prefix(i) for i in range(30)], axis=1)

# 取出df的date列和change列
data_y = df[['date', 'change']].shift(1)

```

```python
data_y.dropna(inplace=True)
```

```python
# 删除data_x的第一行
data_x.drop(data_x.index[0], inplace=True)
```

```python
# 重置data_x的index
data_x.reset_index(drop=True, inplace=True)
data_x
```

```python
data_x.dropna(inplace=True)
data_x
```

```python
# 删掉data_x中的date列
data_x.drop(['0date','1date','2date','3date','4date','5date','6date','7date','8date','9date',
            '10date','11date','12date','13date','14date','15date','16date','17date','18date','19date',
            '20date','21date','22date','23date','24date','25date','26date','27date','28date','29date'], axis=1, inplace=True)
data_x
```

```python
# 把data_y 按照change划分为每隔0.01的指定区间
bins = np.arange(-0.1, 0.11, 0.01)
bins[10] = 0
data_y['change_group'] = pd.cut(data_y['change'], bins)
# 将数据转换为编码
data_y['change_group_code'] = data_y['change_group'].cat.codes
data_y
```

```python
data_y=data_y[:451]
```

```python
data_y.reset_index(drop=True, inplace=True)
data_y
```

```python
data_y.drop(['date'], axis=1, inplace=True)
data_y
```

```python
# 把data_y的change_group列转换为字符串
data_y['change_group'] = data_y['change_group'].astype(str)
data_y['change_group']
```

```python
# 保存data_x和data_y
data_x.to_csv(r'train_test_data/data_x.csv', index=False)
data_y.to_csv(r'train_test_data/data_y.csv', index=False)
```
