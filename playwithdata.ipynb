# 对于feature的处理

```python
# 使用pandas创建两个dataframe
df1 = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range('1/1/2000', periods=1000), columns=list('ABCD'))
df2 = pd.DataFrame(np.random.randn(1000, 4), index=pd.date_range('1/1/2000', periods=1000), columns=list('ABCD'))
```

```python
df1.head()
```

```python
df2.head()
```

```python
# 将df2列名前面加上一个字母E
df2 = df2.add_prefix('E')
df2
```

```python
# 将df1和df2合并
df = pd.concat([df1, df2], axis=1)
```

```python
df
```

```python
# 创建10个df1的平移数据dataframe
df_shift = pd.concat([df.shift(i) for i in range(10)], axis=1)
df_shift.head()
```

```python
# 创建10个df1的平移数据dataframe
df1_shift = pd.concat([df1.shift(i).add_prefix(i) for i in range(10)], axis=1)
df1_shift.head()
```

```python
# 创建数据集
scores = [78, 89, 90, 55, 34, 67, 70, 88, 76, 92, 58, 80, 82, 85, 95, 77, 83, 68, 72, 90]
score_data = pd.DataFrame({'score': scores})

# 将数据划分为指定区间
bins = [0, 59, 79, 100]
score_data['score_group'] = pd.cut(score_data['score'], bins)
```

```python
# 查看score_group的数据类型
score_data['score_group'].dtype
```

```python
import pandas as pd

# 创建数据集
colors = pd.Categorical(['blue', 'red', 'green', 'red', 'blue'])
color_data = pd.DataFrame({'color': colors})
color_data
```

```python

# 将数据转换为编码
color_data['color_code'] = color_data['color'].cat.codes
color_data
```

## 对于label处理的汇总

```python
# 生成-0.1到0.1的随机数
change = pd.DataFrame(np.random.uniform(-0.1, 0.1, size=(20, 1)), index=pd.date_range('1/1/2000', periods=20), columns=["price_change"])
# 将数据change划分为每隔0.01的指定区间
bins = np.arange(-0.1, 0.11, 0.01)
bins[10] = 0
change['price_change_group'] = pd.cut(change['price_change'], bins)
# 将数据转换为编码
change['price_change_group_code'] = change['price_change_group'].cat.codes
change
```
