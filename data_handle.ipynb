## 拼接dataframe

```python

# 创建两个数据框
df1 = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})
df2 = pd.DataFrame({'A': [4, 5, 6], 'B': [7, 8, 9]})

# 按行堆叠数据框
df = pd.concat([df1, df2],axis=1)

# 显示结果
print(df)
```
